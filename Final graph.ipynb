```python
from sklearn.datasets import load_files       
from sklearn.model_selection import train_test_split
# from keras.utils import np_utils
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import normalize
```

```python
size = np.array([700,300,80,60,20])
```

```python
accuracy = np.array([86.4,86.1,86,85.4,84.2])
score = accuracy/size
```

```python
plt.plot(size,score)
plt.title('Score v/s Size of Model')
plt.xlabel('Size of Model')
plt.ylabel('Score')
plt.savefig('Score vs Size of Model.png')
plt.show()
```

```python
plt.plot(size,accuracy)
plt.title('Accuracy v/s Size of Model')
plt.xlabel('Size of Model')
plt.ylabel('Accuracy')
plt.savefig('Accuracy vs Size of Model.png')
plt.show()
```

```python
plt.plot(accuracy, score)
plt.title('Score v/s Accuracy')
plt.xlabel('Accuracy')
plt.ylabel('Score')
plt.savefig('Score vs Accuracy.png')
plt.show()
```
